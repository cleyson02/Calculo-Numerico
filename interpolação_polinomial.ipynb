Qual o calor específico da água a 32.5 ◦C? Construa um programa (em python ou em outra linguagem de sua preferência) que calcula o valor pedido usando a interpolação polinomial e obtenha p2(32.5).

```python
def lagrange_interpolation(x, x_valores, y_valores):
    n = len(x_valores)
    p = 0
    for i in range(n):
        L = 1
        for j in range(n):
            if i != j:
                L *= (x - x_valores[j]) / (x_valores[i] - x_valores[j])
        p += y_valores[i] * L
    return p

# Dados fornecidos na tabela
temperatura = [35, 40, 45]
calor_especifico = [0.99818, 0.99828, 0.99849]

# Valor de x para interpolar
x = 32.5

# Calcula o calor específico usando interpolação polinomial
calor_especifico_32_5 = lagrange_interpolation(x, temperatura, calor_especifico)

# Imprime o resultado
print("O calor específico da água a 32.5°C é aproximadamente:", calor_especifico_32_5)
```

Qual a temperatura tal que o calor específico é 0.99837? Este item pode ser resolvido usando iterpolação linear.

```python
def linear_interpolation(x, x1, y1, x2, y2):
    # Fórmula corrigida da interpolação linear
    y = y1 + (x - x1) * ((y2 - y1) / (x2 - x1))
    return y

# Dados fornecidos na tabela
temperatura = [35, 40, 45]
calor_especifico = [0.99818, 0.99828, 0.99849]

# Valor de y para interpolar
y = 0.99837

# Encontra os pontos mais próximos
x1 = None
y1 = None
x2 = None
y2 = None

for i in range(len(calor_especifico) - 1):
    if calor_especifico[i] <= y <= calor_especifico[i+1]:
        x1 = temperatura[i]
        y1 = calor_especifico[i]
        x2 = temperatura[i+1]
        y2 = calor_especifico[i+1]
        break

# Calcula a temperatura usando interpolação linear
temperatura_interpolada = linear_interpolation(y, y1, x1, y2, x2)

# Imprime o resultado
print("A temperatura correspondente ao calor específico 0.99837 é aproximadamente:", temperatura_interpolada, "°C")
```
