```python
import math

def trapezio(func, a, b, n):
    h = (b - a) / n
    integral = 0.5 * (func(a) + func(b))
    for i in range(1, n):
        x = a + i * h
        integral += func(x)
    integral *= h
    return integral

def func(x):
    return math.exp(-x**2)

a = 0
b = 1
n = 41

aprox_integral = trapezio(func, a, b, n)
print(aprox_integral)
```
