```python
#Método da Bissecção

import math

def f(x):
  return 10*math.log(x) - x**2

#Dados iniciais

#Um intervalo que contém a raíz

a = 1
b = 2

#Precisão

erro = 0.00009

#Contagem de iterações

k = 0

while b-a > erro:
  M = (a + b) / 2

  if f(a)*f(M) < 0:
    b = M

  else:
    a = M

  k = k + 1

print ("A raiz encontrada foi", (a + b) / 2)
print ("O número de iterações", k)
```

```python
#Método da Posição Falsa

import math

def f(x):
  return 10*math.log(x) - x**2

#Dados iniciais

#Um intervalo que contém a raíz

a = 1
b = 2

#Precisão

erro = 0.00009

#Contagem de iterações

k = 0

M = (a + b) / 2

while abs(f(M)) > erro:
  M = (a*f(b) - b*f(a)) / (f(b) - f(a))

  if f(a)*f(M) < 0:
    b = M

  else:
    a = M

  k = k + 1

print("A raiz encontrada foi", M)
print("O número de iterações", k)
```

```python
#Método do Ponto Fixo

import math

#Função original

def f(x):
    return 10*math.log(x) - x**2

#Função de iteração

def phi(x):
    return math.exp((x**2)/10)

#Dados iniciais

#Aproximação inicial

x_k = 1.5

#Precisão

erro = 0.00009

#Contagem de iterações

k = 0

while abs(f(x_k)) > erro:
  x_k = phi(x_k)

  k = k + 1

print("A raiz encontrada foi", x_k)
print("O número de iterações", k)
```

```python
#Método de Newton

import math

def f(x):
    return 10*math.log(x) - x**2

def f_linha(x):
    return 10/x - 2*x

#Dados iniciais

#Aproximação inicial

x_k = 1.5

#Precisão

erro = 0.00009

#Contagem de iterações

k = 0

while abs(f(x_k)) > erro:
    x_k = x_k - f(x_k)/f_linha(x_k)

    k = k + 1

print("A raiz encontrada foi", x_k)
print("O número de iterações", k)
```

```python
#Método da Secante

import math

def f(x):
  return 10*math.log(x) - x**2

#Dados iniciais

#Um intervalo que contém a raíz

x_0 = 1
x_1 = 2

#Precisão

erro = 0.00009

#Contagem de iterações

k = 0

while abs(f(x_0)) > erro:
  x_2 = ((x_0*f(x_1)) - (x_1*f(x_0))) / (f(x_1) - f(x_0))
  x_0 = x_1
  x_1 = x_2

  k = k + 1

print("A raiz encontrada foi", x_2)
print("O número de iterações", k)
```
